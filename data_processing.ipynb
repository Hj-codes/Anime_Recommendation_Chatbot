```python
import pandas as pd
import json

# Load the JSON data
with open('Anime_data.json', "r", encoding="utf-8") as f:
   data = json.load(f)

# Extracting relevant information from the dataset
anime_data = []

for entry in data:
   anime_info = entry['Anime']
   anime_data.append(anime_info)

# Creating dataframe for Anime data
anime_df = pd.DataFrame(anime_data)

# Create a DataFrame from the 'Tags' dictionary
tags_df = anime_df['Tags'].apply(pd.Series)

# Concatenate the original df with the new 'tags_df' DataFrame
anime_df = pd.concat([anime_df.drop(['Tags'], axis=1), tags_df], axis=1)

stats_df = anime_df['stats'].apply(pd.Series)

# Concatenate the original df with the new 'stats_df' DataFrame
anime_df = pd.concat([anime_df.drop(['stats'], axis=1), stats_df], axis=1)

# Replace NaN values with '0%'
anime_df.fillna('0%', inplace=True)

# Remove the percentage sign from 'Average_scores' and convert to int
anime_df['Average_scores'] = anime_df['Average_scores'].str.rstrip('%').astype('int')

# Remove the percentage sign from tag values and convert to int
for col in anime_df.columns:
   if anime_df[col].dtype == 'object' and col not in ['Anime_Title', 'Description', 'stats']:
       anime_df[col] = anime_df[col].str.rstrip('%').astype('int')


# Remove duplicates
anime_df.drop_duplicates(subset=['Anime_Title', 'Description'], inplace=True)

anime_df
```

```python
anime_df.to_csv('anime_data.csv', index=False)
```
